```python
import requests, re
from bs4 import BeautifulSoup

r = requests.get("https://www.century21.ca/search/Q-montreal/45.75035638507768;-74.1508183704102;45.36525235123009;-73.29800464970708/v_Gallery")

cont = r.content

soup = BeautifulSoup(cont,"html.parser")

allprice = soup.find_all("div",{"class":"list-price"})
allbeds = soup.find_all("div",{"class":"basic-stats"})

print(allprice)
print(allbeds)
```

```python
print(allprice[1].find("span",{"class":"price"}).text)
```

```python
l=[]
d={}
for item in allprice:
    try:
        d["price"] = item.find("span",{"class":"price"}).text
    except:
        d["price"] = None

for item in allbeds:
    try:
        d["beds"] = item.find("li",{"class":"beds"}).text
    except:
        d["beds"] = None
    l.append(d)
```
